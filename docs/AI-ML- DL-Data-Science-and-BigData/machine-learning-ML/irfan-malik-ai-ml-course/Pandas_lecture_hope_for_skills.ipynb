**Data Types**

Data types bohat zaroori hain kyun k ye computer ko batate hain ke data ko kis tarah se samjhein aur us par kya amal karein. Data types har programming language mein mojood hote hain aur har qisam ke data ko sahi tareeqe se store aur process karne ke liye zaroori hain.

Data types ki madad se hum ye bhi yakeen kar sakte hain ke jo data hum collect kar rahe hain wo sahi format aur value mein hai. Jaise agar hum kisi ka naam store karna chahte hain to hum us ko string data type mein rakhenge, aur agar hum kisi ka phone number store karna chahte hain to hum us ko integer data type mein rakhenge.

Data types ke bohat se qisam hain, jaise integer, float, string, boolean, date, time, array, etc. Har data type ka apna maqsad aur istemal hai. Jaise integer data type sirf mukammal numbers ko store karta hai, float data type decimal numbers ko store karta hai, string data type text ya characters ko store karta hai, boolean data type sirf true ya false values ko store karta hai, date data type tareekh ko store karta hai, time data type waqt ko store karta hai, array data type aik list ya silsila ko store karta hai, etc

## --int64 <br>
## --object <br>
##--float64

-- **int64**: int64 aik qisam ka data hai jo sirf mukammal numbers ko store kar sakta hai. Mukammal numbers wo hain jo koi decimal point ya fraction na hon, jaise 1, 2, 3, … Int64 ka matlab hai ke is data type mein aik number ko store karne ke liye 64 bits ki zaroorat hoti hai. Aik bit aik chota sa switch hai jo sirf do states mein ho sakta hai: on ya off, ya 1 ya 0. Int64 mein hum 2^64 mukhtalif numbers ko store kar sakte hain.

-- **object**: object aik qisam ka data hai jo kisi bhi cheez ko store kar sakta hai. Object mein hum text, numbers, dates, ya koi bhi aur data rakh sakte hain. Object flexible hai lekin is ki efficiency kam hoti hai. Object ko hum container keh sakte hain jo kuch bhi rakh sakta hai.

-- **float64**: float64 aik qisam ka data hai jo decimal point ya fraction wale numbers ko store kar sakta hai. Jaise 1.5, 3.14, 0.25, … Float64 ka matlab hai ke is data type mein aik number ko store karne ke liye 64 bits ki zaroorat hoti hai. Float64 mein hum bohat saare decimal numbers ko store kar sakte hain lekin un ki accuracy poori nahi hoti.
